```python
online = pd.read_csv("thetaPSD_online_2back_-1580742246_20260602_100603.csv")
online
```

```python
theta_z_csv = pd.read_csv("theta_z_values_2back_-1580742246.csv", header= None, names = ["Time", "online z_score"])
theta_z_csv["Time"] = theta_z_csv["Time"] - theta_z_csv["Time"].iloc[0]
theta_z_csv
```

```python
merged_df = pd.merge_asof(
    left=online.sort_values("Time"),
    right=theta_z_csv.sort_values("Time"),
    on="Time",
    direction="nearest",
    tolerance=0.005  # optional: max allowed difference in seconds
)
merged_df
```

```python
merged_df[" Ch04"][0]
```

```python
buffer = []
theta = []

SONICATION_TIME = 3
COOLDOWN_TIME = 5
THETA_THRESHOLD_Z = 1.5
MU = 2.32
SIGMA = 4.18
MAD_THRESHOLD = 6
INITIAL_CUTOFF = 100.0
BUFFER_SIZE = 500

for i in range(len(merged_df)):

    theta_val = merged_df[" Ch04"][i]

    if len(buffer) <= 450:
        if theta_val < INITIAL_CUTOFF:
            buffer.append(theta_val)
        theta.append(np.nan)     # keep alignment
        continue

    if len(buffer) > BUFFER_SIZE:
        buffer.pop(0)

    arr = np.array(buffer)
    median = np.median(arr)
    mad = np.median(np.abs(arr - median)) + 1e-6

    z_art = abs(theta_val - median) / mad
    if z_art > MAD_THRESHOLD:
        theta.append(np.nan)
        continue

    buffer.append(theta_val)
    theta_z = abs(theta_val - MU) / SIGMA
    theta.append(theta_z)
```

```python
merged_df["offline z-score"] = theta
merged_df["online z_score"] = merged_df["online z_score"].round(2)
merged_df["offline z-score"] = merged_df["offline z-score"].round(2)
merged_df
```
